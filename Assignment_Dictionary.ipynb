```python
#create a dictionary to hold the K:V pairs
myItems = {}
#First step is to load the file into your program.
file_path = """C:\SEIS603_PYTHON\RetailStoreItemData.txt"""
with open(file_path, "r") as file: #read items in
    header_line = next(file)  # Im storing the header if I need it later
#You now need to only parse the description and the price 
#and save both as key and value in a dictionary object called myItems.
    for line in file:
        #split the line by commas
        parts = line.split(',')
        # get the "Description" and "Unit price" which are [1] and [6] in the index
        # Assuming the structure is consistent with the sample, where Description is the second column, and Unit price is the seventh
        description = parts[1].strip()  # Remove lead/trailing spaces
        unit_price = parts[6].strip() # Remove lead/trailing spaces
        myItems[description] = unit_price # Assign a value(unit_price) to a key(description)
#print(header_line)
#show some values to make sure it works
#list(myItems.items())[:]
```

```python
#Part 2
#Ask user to enter a description of an item and when they provide than go search 
#that in your dictionary object and display the description as 
#well as the associated price and ask the user how many would they like to buy? 
#so we need another dictionary to store items we want to store in the cart
mycart = {}
#sample items Banana and Orange
#correction. there are no orange in teh txt file
while True:
    # Ask the user what item they would like to buy
    description = input("What item would you like to buy? Or select x to exit: ")
    #f the user selects x then simply do the final calculations and show the results as follows.
    #You should plan to use 7% for tax calculation
    if description.lower() == 'x': 
        # User chose to exit; calculate and display final total
        subtotal = sum(mycart.values()) #take the unit prices collected from dictionary
        #and sum them
        tax = subtotal * 0.07  # 7% tax
        total = subtotal + tax
        print("Your subtotal: ${:.2f}".format(subtotal))
        print("Tax: ${:.2f}".format(tax))
        print("Total: ${:.2f}".format(total))
        break
    else:
        # search for the item in myItems and display the price
        if description in myItems:
            unit_price = float(myItems[description])
            print("{} ${} each".format(description, unit_price))
            
            # Ask how many they would like to buy
            quantity = input("How many {} would you like to buy? ".format(description))
            quantity = int(quantity)  # Convert quantity to an integer
            
            # calc. the total for this item and add to the shopping cart
            total_price = unit_price * quantity
            mycart[description] = total_price
            print("Item added to your cart.")
        else:
            print("Item not found. Please try again.")
```
